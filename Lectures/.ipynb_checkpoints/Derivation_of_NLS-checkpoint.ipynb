# The Derivation of the Nonlinear Schroedinger Equation

We start from the non-dimensionalized, infinite depth water wave problem in the form 

\begin{align*}
\Delta \phi = 0, & ~-\infty < z < \epsilon \eta(x,t), \\
\phi_{z} \rightarrow 0, & ~z \rightarrow -\infty,\\
\eta_{t} + \epsilon \eta_{x}\phi_{x} - \phi_{z} = 0, & ~ z = \epsilon \eta(x,t),\\
\phi_{t} + \frac{\epsilon}{2}\left(\phi_{x}^{2} + \phi_{z}^{2}\right) + \eta = 0, &~ z = \epsilon \eta(x,t).
\end{align*}


## The Linear Problem 
Using Fourier transforms, we can solve Laplace's equation in a relatively straightforward way where 
$$
-k^{2}\hat{\phi} + \hat{\phi}_{zz} = 0,
$$
so, taking the boundary condition into account and then inverting, we find the potential is given by 
$$
\phi(x,z,t) = \frac{1}{2\pi}\int_{\mathbb{R}}e^{ikx + |k|z}\hat{A}(k,t) dk + \mbox{c.c.}
$$
Linearizing the free-boundary conditions, which we can do by setting $\epsilon=0$, we readily find 
$$
\hat{A}(k,t) = e^{-i\omega(k)t}\hat{A}(k),
$$
where the dispersion relationship $\omega(k) = \pm \sqrt{|k|} = \pm \sqrt{sk}$, where $s = \mbox{sgn}(k)$.

## Surface Variable Formulation and Dirichlet--Neumann Operators 

We define the surface velocity potential $q(x,t) = \phi(x,\epsilon \eta(x,t), t)$.  We readily see that 
$$
q_{x} = \phi_{x} + \epsilon \eta_{x}\phi_{z},  
$$
so that, using the first free boundary equation, we have the following linear system 
$$
\begin{pmatrix}q_{x} \\ \eta_{t}\end{pmatrix} = \begin{pmatrix}1 & \epsilon \eta_{x}\\ -\epsilon \eta_{x} & 1\end{pmatrix}\begin{pmatrix}\phi_{x} \\ \phi_{z}\end{pmatrix}.
$$
Inverting and substituting into the second nonlinear-free-surface equation, using
$$
q_{t} = \epsilon \eta_{t}\phi_{z} + \phi_{t},
$$
and then expanding up to $\mathcal{O}(\epsilon^{2})$ terms, we get the equation 
$$
q_{t} + \eta + \epsilon \left(\frac{1}{2}\left(q_{x}^{2} - \eta_{t}^{2} \right)-\epsilon \eta_{t}\eta_{x}q_{x} \right) = 0.
$$

We are now left with the question of what to do with the first nonlinear-free-surface equation.  To address this, we see that 
\begin{align}
q(x,t) = & \frac{1}{2\pi}\int_{\mathbb{R}}\left(\hat{A}(k,t)e^{ikx} + \hat{A}^{\ast}(k,t)e^{-ikx}\right) e^{\epsilon|k|\eta(x,t)}dk\\
 = & \frac{1}{2\pi}\int_{\mathbb{R}}\left(\hat{A}(k,t) + \hat{A}^{\ast}(-k,t)\right)e^{ikx} e^{\epsilon|k|\eta(x,t)}dk\\
 = & \frac{1}{2\pi}\int_{\mathbb{R}} \hat{\tilde{A}}(k,t) e^{ikx} e^{\epsilon|k|\eta(x,t)}dk, ~ \mbox{where}~\hat{\tilde{A}}(k,t) = \hat{A}(k,t) + \hat{A}^{\ast}(-k,t)\\
 = & \frac{1}{2\pi}\int_{\mathbb{R}} \left(1 + \epsilon |k|\eta(x,t) + \frac{\epsilon^{2}\eta^{2}(x,t)}{2}|k|^{2}+\cdots\right)\hat{\tilde{A}}(k,t) e^{ikx} dk\\
 = & \tilde{A} - \epsilon \eta \mathcal{H}\tilde{A}_{x} - \frac{\epsilon^{2}\eta^{2}}{2}\tilde{A}_{xx} + \cdots 
\end{align}
where 
$$
\tilde{A}(x,t) = \frac{1}{2\pi}\int_{\mathbb{R}}\hat{\tilde{A}}(k,t)e^{ikx}dk
$$
and the _Hilbert transform_ $\mathcal{H}$ is given via the formula 
$$
\mathcal{H}f(x) = \frac{1}{2\pi}\int_{\mathbb{R}}i\mbox{sgn}(k) \hat{f}(k) e^{ikx}dk.  
$$
Here the _symbol_ associated with $\mathcal{H}$ is $i\mbox{sgn}(k)$.  This definition through the Fourier transform makes the Hilbert transform what is called a _pseudo-differential operator_.  

Given the correspondance we now have between $q$ and $\tilde{A}$, we can invert this by letting 
$$
\tilde{A} = q + \epsilon R_{1} + \epsilon^{2}R_{2} + \cdots 
$$
so that 
$$
\epsilon\left(R_{1} -\eta \mathcal{H}\left(\tilde{q}_{x}+\epsilon \partial_{x}R_{1}\right)\right) +\epsilon^{2}\left(R_{2}- \frac{\eta^{2}}{2}\tilde{q}_{xx}\right) + \cdots = 0.
$$
Matching orders of $\epsilon$ then gives us 
$$
R_{1} = \eta \mathcal{H}q_{x}, ~ R_{2} = \eta \mathcal{H}\partial_{x}R_{1} + \frac{\eta^{2}}{2}q_{xx}
$$

Now, in order to rewrite the equation 
$$
\eta_{t} = -\eta_{x}\phi_{x} + \phi_{z}, ~ z = \epsilon \eta(x,t),
$$
we must expand the integrals
$$
\phi_{x}(x,\epsilon\eta,t) = \frac{1}{2\pi}\int_{\mathbb{R}}ik\hat{\tilde{A}}(k,t)e^{ikx}e^{\epsilon|k|\eta}dk,
$$
and
$$
\phi_{z}(x,\epsilon\eta,t) = \frac{1}{2\pi}\int_{\mathbb{R}}|k|\hat{\tilde{A}}(k,t)e^{ikx}e^{\epsilon|k|\eta}dk.
$$
Expanding to the appropriate powers of $\epsilon$, we have that 
$$
\phi_{x}(x,\epsilon\eta,t) = \tilde{A}_{x} - \epsilon \eta\mathcal{H}\tilde{A}_{xx} + \cdots,
$$
and
$$
\phi_{z}(x,\epsilon\eta,t) = -\mathcal{H}\tilde{A}_{x} - \epsilon \eta\tilde{A}_{xx} + \frac{\epsilon^{2}\eta^{2}}{2}\mathcal{H}\tilde{A}_{xxx} + \cdots,
$$
Substituting and collecting powers we then find that 
\begin{align}
\eta_{t} = & -\mathcal{H}q_{x} + \epsilon\left(-\mathcal{H}\partial_{x}R_{1} - \partial_{x}\left(\eta q_{x}\right)\right) \\
 & + \epsilon^{2}\left(-\mathcal{H}\partial_{x}R_{2}- \partial_{x}\left(\eta\partial_{x}R_{1}\right) + \partial_{x}\left( \frac{\eta^{2}}{2}\mathcal{H}q_{xx}\right)\right) + \cdots 
\end{align}

## Slow Modulation Approximations and the NLS Equation
From above, we now need to find an approximate solution to the coupled nonlinear system of equations
\begin{align}
\eta_{t} = & -\mathcal{H}Q + \epsilon\left(-\mathcal{H}\partial_{x}R_{1} - \partial_{x}\left(\eta Q\right)\right) \\
 & + \epsilon^{2}\left(-\mathcal{H}\partial_{x}R_{2}- \partial_{x}\left(\eta\partial_{x}R_{1}\right) + \partial_{x}\left( \frac{\eta^{2}}{2}\mathcal{H}Q_{x}\right)\right) + \cdots 
\end{align}
and
$$
Q_{t} + \eta_{x} + \epsilon\partial_{x} \left(\frac{1}{2}\left(Q^{2} - \eta_{t}^{2} \right)-\epsilon \eta_{t}\eta_{x}Q \right) = 0.
$$
where
$$
R_{1} = \eta \mathcal{H}Q, ~ R_{2} = \eta \mathcal{H}\partial_{x}R_{1} + \frac{\eta^{2}}{2}Q_{x}
$$
and where $Q=q_{x}$, the introduction of which simplifies further computations.  

In order to proceed further, we introduce a _slow modulation_ approximation, whereby we suppose that 
$$
Q(x,t) = \frac{1}{2\pi \epsilon}\int_{\mathbb{R}}\hat{Q}\left(\frac{k-k_{0}}{\epsilon} \right)e^{i(kx-\omega(k)t)}dk + \mbox{c.c.},
$$
with the idea that $\left|\hat{Q}(k)\right|\rightarrow 0$ rapidly as $|k|\rightarrow \infty$.  Thus, we suppose that the solution $Q(x,t)$ is narrowly banded around the carrier wavenumber $k_{0}$.  Transforming via $\tilde{k}=(k-k_{0})/\epsilon$ and expanding in $\tilde{k}$, we have, after dropping tildes, that 
$$
Q(x,t) \approx \frac{e^{i\theta(x,t,k_{0})}}{2\pi}\int_{\mathbb{R}}\hat{Q}\left(k\right)e^{i(k\xi-k^{2}\omega''(k_{0})\tau/2)}dk + \mbox{c.c.},
$$
where
$$
\theta(x,t,k_{0}) = k_{0}x-\omega(k_{0})t, ~ \xi = \epsilon(x-c_{g}(k_{0})t), ~ \tau = \epsilon^{2}t.
$$
